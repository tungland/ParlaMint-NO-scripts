```python
import json
import re
import pandas as pd
```

```python
lst = list()
with open('search_result.jl') as f:
    for line in f.readlines():
        lst += [json.loads(line)]
```

```python
new_dct_lst = dict()
new_dct_str = dict()
for x in lst:
    #print(x.keys())
    for key, val in x.items():
        #
        #new_dct[key] = json.loads(val)
        #print(type(key), type(val))
        print(key, type(key))
        print( type(val))
        
        if isinstance(val, str):
            new_dct_str[key] = json.loads(val)
        else:
            new_dct_lst[key] = val
```

```python
link_dct = dict()
for key, val in new_dct_str.items():
    link_dct[key] = [e['link']  for e in val['organic_results']]
```

```python
dct = {}
dct.update(link_dct)
dct.update(new_dct_lst)
```

```python
p = r"https:\/\/www.stortinget\.no\/no\/Representanter-og-komiteer\/Representantene\/Representant\/\?perid=([A-ZÆØÅ]{2,5})(.+)?"


dct_matches = dict()
for key, val in dct.items():
   dct_matches[key] =  set([re.match(p, r).group(1)  for r in val if re.match(p, r)])
    
```

```python
# Check for set longer than 1
for key, val in dct_matches.items():
    if len(val) > 1:
        raise ValueError
    
    elif len(val) == 1:
        dct_matches[key] = list(val)[0]
        
    else:
        dct_matches[key] = None
```

```python
df = pd.DataFrame.from_dict(dct_matches, orient='index').reset_index()
df
```
