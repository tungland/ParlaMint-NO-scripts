```python
import lxml.etree as et
import pandas as pd

from tqdm import tqdm
from glob import glob
```

```python
docs = glob('../../output/*')
```

```python
xml = et.parse(docs[0])
```

```python
dct = dict()

for doc in tqdm(docs):
    xml = et.parse(doc)
    
    #for u in xml.findall("/./{http://www.tei-c.org/ns/1.0}text/{http://www.tei-c.org/ns/1.0}body/{http://www.tei-c.org/ns/1.0}div/{http://www.tei-c.org/ns/1.0}u"):
    for u in xml.findall("//{http://www.tei-c.org/ns/1.0}u"):
        if '{http://www.w3.org/XML/1998/namespace}lang' in u.attrib:    
            #if u.attrib['{http://www.w3.org/XML/1998/namespace}lang'] not in ['nno', 'nob']:
            dct.update(
                {u.attrib['{http://www.w3.org/XML/1998/namespace}id'] : {
                                                'who' : u.attrib['who'],
                                                'lang' : u.attrib['{http://www.w3.org/XML/1998/namespace}lang'],
                                                'text' : u.xpath('tei:seg/text()', namespaces={'tei' : 'http://www.tei-c.org/ns/1.0'})
                                                }
                    }
                )       
dct
```

```python
u.attrib
```

```python
df = pd.DataFrame.from_dict(dct, orient='index')

df
```

```python
t = df.groupby(['who', 'lang']).count()
```
