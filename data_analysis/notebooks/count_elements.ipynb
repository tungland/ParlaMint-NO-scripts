```python
import glob
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
session_paths = glob.glob("data/*")
sessions = [s.split("/")[1] for s in session_paths]
```

```python
path_dct = {}
for s in session_paths:
    path_dct[s.split("/")[1]] = glob.glob(s + "/*")
```

```python
def count_elements(xml_list):
    dct = {}
    
    
    for doc in xml_list: 
        with open(doc) as f:
            soup = bs(f, "lxml-xml")   
        for desc in soup.descendants:
            if desc.name in dct.keys():
                dct[desc.name] += 1
            else:
                dct[desc.name] = 1
    return dct
```

```python
dct = {}
for s in sessions:
    inner_dct = count_elements(path_dct[s])
    dct[s] = inner_dct
```

```python
pd.DataFrame(dct).transpose().sort_index(ascending=False)
```
