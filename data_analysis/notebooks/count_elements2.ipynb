```python
import glob
import pandas as pd
import numpy as np
```

```python
c = glob.glob("elements_data/*")
```

```python
df_dct = {}

for csv in c:
    df_dct.update({csv.split("/")[1].split(".")[0] : pd.read_csv(csv)})
```

```python
sessions = sorted(list(df_dct.keys()))
```

```python
dct = {}
for s in sessions:
    if '0' in df_dct[s].columns:
        dct.update({s : df_dct[s].sort_values('0', ascending=False).dropna().iloc[:, 0].reset_index(drop=True)})
```

```python
df = pd.DataFrame(dct)#.transpose()
df
```

```python
for r in (df['2006-2007']):
    print(r)
```

```python
dct = {}
for s in sessions:
    if '0' in df_dct[s].columns:
        dct.update({s : df_dct[s].sort_values('0', ascending=False).dropna().iloc[:, 0].reset_index(drop=True)})
```

```python
df = pd.DataFrame(dct).transpose()
```

```python
n_dct = {}
for row in df.iterrows():
    n_dct[row[0]] = [x for x in row[1] if x is not np.nan]
```

```python
n_dct
```

```python
lst = []
for row in df.iterrows():

    lst += [x for x in row[1] if x is not np.nan]
```

```python
df = pd.DataFrame(lst)
```

```python
d = df[0].apply(lambda x : x.split(" "))
```

```python
df[0]
```

```python
dct = {}
for doc in d:
    for word in doc:
       if word in dct.keys():
           dct[word] += 1
       else:
           dct[word] = 1
```

```python
df
```

```python
df = pd.DataFrame.from_dict(dct, orient = 'index')
```

```python
df.sort_values(0, ascending=False).to_csv("element_count.csv")
```
