```python
import glob
import pandas as pd
import numpy as np
```

```python
c = glob.glob("elements_data/*")

df_dct = {}

for csv in c:
    df_dct.update({csv.split("/")[1].split(".")[0] : pd.read_csv(csv)})
    
sessions = sorted(list(df_dct.keys()))
```

```python
dct = {}
for s in sessions:
    if '0' in df_dct[s].columns:
        dct.update({s : df_dct[s].sort_values('0', ascending=False).dropna().iloc[:, 0].head(15).reset_index(drop=True)})
```

```python
df = pd.DataFrame(dct)
```
